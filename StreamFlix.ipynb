```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"Movies_clean.csv")

print("Data Head")
print(data.head())
```

```python
print("Data isna")
print(data.isna().sum())
```

```python
print("Data Duplicated")
print(data.duplicated().sum())
```

```python
print("Data Describe")
print(data.describe())
```

```python
print("Data Heatmap")
sns.heatmap(data.isnull(), cmap="coolwarm", cbar=False)
plt.title("Missing Data Heatmap")
plt.show()
```
